```python
import numpy as np
import scipy as sp
from matplotlib import pyplot as plt
```

```python
import Metric as mt
import Calculus as cl
import Geodesics as gd
import Constants as cte
import Christoffel as ch
```

```python
def Potential(r, epsilon, L):
    return (0.5*epsilon - 0.5*cte.Rs*epsilon/r + 0.5*L**2/(r**2) - 0.5*cte.Rs*L**2/(r**3))
```

```python
t = 0
r0 = 5*cte.Rs
T0 = 20*cl.dt
L0 = (2*np.pi*r0**2/(8e4*T0))
E0 = 1

x0 = np.array([r0,0.5*np.pi,0, 0])
x1 = np.array([E0-Potential(r0,0.01, L0), 0, L0/r0**2, E0/mt.g_ij(x0)[0,0]])
x2 = np.array([-cte.Rs*0.5/(r0*r0),0,0,0])
```

```python
r0 = np.linspace(0.01,r0)
V = Potential(r0, 0.01, L0)
plt.plot(r0,V)
plt.show()
```

```python
%%time
data = gd.SolveGeodesics(mt.g_ij, mt.gij_, x0, x1, x2, t,5000)
```

```python
#Graficar x, y . La orbita
x_position = data[:,9]
y_position = data[:,10]
plt.plot(x_position,y_position)
plt.grid(True, which='both')
plt.show()
```
